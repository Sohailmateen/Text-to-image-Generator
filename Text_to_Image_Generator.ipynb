```python
from diffusers import StableDiffusionPipeline
import torch

# Load the model with FP16 precision for GPU
model = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)
model = model.to("cuda")  # Move the model to GPU
```

```python
prompt = "a beautiful beach during sunset"  # Enter your desired text prompt
image = model(prompt).images[0]  # Generate the image
image.save("generated_image.png")  # Save the image locally
image.show()  # Display the image in the notebook
```
